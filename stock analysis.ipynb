```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("stocks.csv")
```

```python
print(data.head())
```

```python
print(data.isnull().sum().sum())
```

```python
print(data.describe())
```

```python
data.dropna(inplace=True)
```

```python
data["Date"]=pd.to_datetime(data["Date"])
```

```python
data.sort_values("Date",inplace=True)
```

```python
data[["Open","High","Low","Close","Adj Close","Volume"]].describe()
```

```python
data[["Open","High","Low","Close","Adj Close"]].hist(figsize=(10,8))
```

```python
import seaborn as sns
sns.boxplot(data=data[["Open","High","Low","Close","Adj Close"]])
```

```python
sns.heatmap(data[["Open","High","Low","Close","Adj Close","Volume"]].corr(),annot=True,cmap="coolwarm")
```

```python
data["price range"]=data["High"]-data["Low"]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(data["Date"],data["Close"])
plt.title("Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
print(data["Date"].dtype)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.bar(data["Date"],data["Volume"],color="lightblue")
plt.title ("stock trading volume over time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(data["Date"],data["Open"],label="Open Price",color="green")
plt.plot(data["Date"],data["Close"],label="Close Price",color="blue")
plt.title("stock price movement(open vs close)over time")
plt.xlabel("date")
plt.ylabel("price")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# the x-axis will show the date of each stock trading day
# the y-axis will show the price for both open and close price
# the green line represents the open price , and the blue line represents the close price
```
